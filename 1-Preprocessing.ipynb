```python
import pandas as pd
import re
import demoji
```

```python
df = pd.read_csv('dogecoin_raw.csv')
df.columns
```

```python
df = df[['date','tweet']]
```

```python
df[:50]
```

```python
def clean_text(tweet):
        """
        Removed hashtags, mentions, multiple spaces and non-ascii
            characters.
        """
        # Remove hashtags
        # tweet = re.sub("#\w*", "", tweet)
        # Remove @
        tweet = re.sub("@\w*", "", tweet)
        tweet = re.sub("&amp", "", tweet)
        tweet = re.sub(r'http\S+', '', tweet)
        tweet = tweet.replace("\n", " ")
        tweet = demoji.replace_with_desc(tweet)
        tweet = tweet.replace(":", " ")
        tweet = re.sub("\s{2,}", " ", tweet)
        # Remove non-ascii
        tweet = tweet.encode("ascii", "ignore").decode("ascii")
        tweet = tweet.strip()
        return tweet
```

```python
df['tweet'] = df['tweet'].apply(lambda x: clean_text(str(x)))
```

```python
df = df.sort_values('date')
```
